```python
import numpy as np
import lsst.daf.butler as dafButler
```

```python
config = 'dp02'
collections = '2.2i/runs/DP0.2'
butler = dafButler.Butler(config, collections=collections)
```

Check the `diaSourceTable`.

```python
datasetType = 'diaSourceTable'
dataId = {'visit': 192350, 'detector': 175}
diaSrc = butler.get(datasetType, dataId=dataId)
print(f"Retrieved catalog of {len(diaSrc)} sources.")
```

```python
for word in ['time', 'Time', 'tai', 'Tai', 'date']:
    for col in diaSrc.columns:
        tmp = str(col)
        if tmp.find(word) > -1:
            print(col)
```
